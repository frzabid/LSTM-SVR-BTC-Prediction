```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.fft import fft, ifft, fftfreq

def vmd(signal, alpha, tau, K, DC, init, tol, fs=1.0, N_iter=500, verbose=False):
    T = len(signal)
    freqs = fftfreq(T, d=1/fs)
    f_hat = fft(signal)

    u_hat = np.zeros((K, T), dtype=complex)
    omega = np.linspace(0, 0.5, K) if init == 1 else np.zeros(K)
    lambda_hat = np.zeros(T, dtype=complex)

    u_diff = tol + 1
    n = 0
    convergence_log = []

    while u_diff > tol and n < N_iter:
        u_hat_old = np.copy(u_hat)

        for k in range(K):
            sum_uk = np.sum(u_hat, axis=0) - u_hat[k, :]
            residual = f_hat - sum_uk - lambda_hat / 2
            denominator = 1 + 2 * alpha * (freqs - omega[k])**2
            u_hat[k, :] = residual / denominator

            if not DC or k != 0:
                power = np.abs(u_hat[k, :])**2
                omega[k] = np.sum(freqs * power) / (np.sum(power) + 1e-12)

        lambda_hat += tau * (np.sum(u_hat, axis=0) - f_hat)
        u_diff = np.sum(np.abs(u_hat - u_hat_old)**2) / K
        convergence_log.append(u_diff)
        if verbose:
            print(f"Iteration {n+1}, u_diff: {u_diff:.2e}")
        n += 1

    u = np.real(ifft(u_hat, axis=1))
    return u, omega, convergence_log

def plot_modes(modes, omega, title="VMD Decomposition"):
    K = modes.shape[0]
    plt.figure(figsize=(14, 2.5 * K))
    for i in range(K):
        plt.subplot(K, 1, i + 1)
        plt.plot(modes[i])
        plt.title(f"Mode {i+1} - Center frequency: {omega[i]:.4f} cycles/sample")
        plt.tight_layout()
    plt.suptitle(title, y=1.02)
    plt.show()

signal = df_Close['BTC_Close_BTC'].astype(float).values
signal = (signal - np.min(signal)) / (np.max(signal) - np.min(signal))

params = {
    "alpha": 2000,
    "tau": 0.,
    "K": 8,
    "DC": 0,
    "init": 1,
    "tol": 1e-7,
    "fs": 1.0,
    "N_iter": 500,
    "verbose": True
}

modes, omega, log = vmd(signal, **params)

print("\n=== Center Frequencies per Mode")
for i, freq in enumerate(omega):
    print(f"Mode {i+1}: {freq:.4f} cycles/sample")

plot_modes(modes, omega, title="VMD Decomposition of BTC Close Price")

modes_df = pd.DataFrame(modes.T, columns=[f'Mode_{i+1}' for i in range(modes.shape[0])])
output_filename = 'vmd_results.csv'
modes_df.to_csv(output_filename, index=False)
print(f"\nResults saved to {output_filename}")

convergence_df = pd.DataFrame({'Convergence_Log': log})
convergence_df.to_csv('vmd_convergence_log.csv', index=False)
print(f"Convergence log saved to vmd_convergence_log.csv")
```

```python
import numpy as np
import math
from collections import defaultdict
import pandas as pd
from sklearn.metrics import mean_absolute_error, r2_score
from sklearn.preprocessing import StandardScaler
from keras.models import save_model
import pickle
import os

class LSTMCell:
    def __init__(self, input_size, hidden_size):
        # Initialize parameters
        self.input_size = input_size
        self.hidden_size = hidden_size
        
        # Initialize weights and biases
        scale = 1.0 / math.sqrt(hidden_size)
        self.Wf = np.random.randn(hidden_size, input_size + hidden_size) * scale
        self.Wi = np.random.randn(hidden_size, input_size + hidden_size) * scale
        self.Wc = np.random.randn(hidden_size, input_size + hidden_size) * scale
        self.Wo = np.random.randn(hidden_size, input_size + hidden_size) * scale
        
        self.bf = np.zeros((hidden_size, 1))
        self.bi = np.zeros((hidden_size, 1))
        self.bc = np.zeros((hidden_size, 1))
        self.bo = np.zeros((hidden_size, 1))
        
    def forward(self, x, h_prev, c_prev):
        # Concatenate input and previous hidden state
        concat = np.vstack((h_prev, x))
        
        # Forget gate
        ft = self.sigmoid(np.dot(self.Wf, concat) + self.bf)
        
        # Input gate
        it = self.sigmoid(np.dot(self.Wi, concat) + self.bi)
        
        # Candidate cell state
        cct = np.tanh(np.dot(self.Wc, concat) + self.bc)
        
        # New cell state
        ct = ft * c_prev + it * cct
        
        # Output gate
        ot = self.sigmoid(np.dot(self.Wo, concat) + self.bo)
        
        # New hidden state
        ht = ot * np.tanh(ct)
        
        return ht, ct, (ft, it, cct, ot, concat)
    
    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))


class BiLSTM:
    def __init__(self, input_size, hidden_size):
        self.input_size = input_size
        self.hidden_size = hidden_size
        
        # Forward and backward LSTM cells
        self.forward_lstm = LSTMCell(input_size, hidden_size)
        self.backward_lstm = LSTMCell(input_size, hidden_size)
        
        # Output layer weights
        self.Wy = np.random.randn(1, 2 * hidden_size) * 0.01
        self.by = np.zeros((1, 1))
        
        # Initialize StandardScaler
        self.scaler = StandardScaler()
        self.fitted_scaler = False
        
    def forward(self, X):
        # Initialize hidden states and cell states
        hf_prev = np.zeros((self.hidden_size, 1))
        cf_prev = np.zeros((self.hidden_size, 1))
        hb_prev = np.zeros((self.hidden_size, 1))
        cb_prev = np.zeros((self.hidden_size, 1))
        
        T = len(X)  # Sequence length
        self.cache = defaultdict(list)
        
        # Forward pass
        h_forward = []
        for t in range(T):
            x = X[t].reshape(-1, 1)
            hf_prev, cf_prev, cache = self.forward_lstm.forward(x, hf_prev, cf_prev)
            h_forward.append(hf_prev)
            self.cache['forward'].append(cache)
        
        # Backward pass
        h_backward = []
        for t in reversed(range(T)):
            x = X[t].reshape(-1, 1)
            hb_prev, cb_prev, cache = self.backward_lstm.forward(x, hb_prev, cb_prev)
            h_backward.insert(0, hb_prev)
            self.cache['backward'].append(cache)
        
        # Combine forward and backward hidden states
        h_combined = [np.vstack((hf, hb)) for hf, hb in zip(h_forward, h_backward)]
        
        # Output prediction
        y_pred = np.dot(self.Wy, h_combined[-1]) + self.by
        
        return y_pred, h_combined
    
    def backward(self, X, y_true, y_pred, h_combined, learning_rate=0.005):
        # Calculate gradient of output
        dy = y_pred - y_true
        
        # Initialize gradients
        dWy = np.dot(dy, h_combined[-1].T)
        dby = dy
        
        # Initialize hidden state gradients
        dh_next_f = np.zeros((self.hidden_size, 1))
        dc_next_f = np.zeros((self.hidden_size, 1))
        dh_next_b = np.zeros((self.hidden_size, 1))
        dc_next_b = np.zeros((self.hidden_size, 1))
        
        # Backward pass through time
        dWf_f, dWi_f, dWc_f, dWo_f = 0, 0, 0, 0
        dbf_f, dbi_f, dbc_f, dbo_f = 0, 0, 0, 0
        
        dWf_b, dWi_b, dWc_b, dWo_b = 0, 0, 0, 0
        dbf_b, dbi_b, dbc_b, dbo_b = 0, 0, 0, 0
        
        T = len(X)
        
        # Backward pass for forward LSTM
        for t in reversed(range(T)):
            # Get cached values
            ft, it, cct, ot, concat = self.cache['forward'][t]
            
            # Gradient from output and next time step
            dh = np.dot(self.Wy[:, :self.hidden_size].T, dy) + dh_next_f
            do = dh * np.tanh(h_combined[t][:self.hidden_size]) * ot * (1 - ot)
            dc = dh * ot * (1 - np.tanh(h_combined[t][:self.hidden_size])**2) + dc_next_f
            dit = dc * cct * it * (1 - it)
            dft = dc * self.cache['forward'][t-1][1] if t > 0 else dc * 0
            dft = dft * ft * (1 - ft)
            dcct = dc * it * (1 - cct**2)
            
            # Update parameter gradients
            dWf_f += np.dot(dft, concat.T)
            dWi_f += np.dot(dit, concat.T)
            dWc_f += np.dot(dcct, concat.T)
            dWo_f += np.dot(do, concat.T)
            
            dbf_f += dft
            dbi_f += dit
            dbc_f += dcct
            dbo_f += do
            
            # Compute gradients for next step
            dconcat = (np.dot(self.forward_lstm.Wf.T, dft) + 
                      np.dot(self.forward_lstm.Wi.T, dit) + 
                      np.dot(self.forward_lstm.Wc.T, dcct) + 
                      np.dot(self.forward_lstm.Wo.T, do))
            
            dh_next_f = dconcat[:self.hidden_size]
            dc_next_f = dc * ft
            
        # Backward pass for backward LSTM
        for t in range(T):
            # Get cached values
            ft, it, cct, ot, concat = self.cache['backward'][t]
            
            # Gradient from output and next time step
            dh = np.dot(self.Wy[:, self.hidden_size:].T, dy) + dh_next_b
            do = dh * np.tanh(h_combined[t][self.hidden_size:]) * ot * (1 - ot)
            dc = dh * ot * (1 - np.tanh(h_combined[t][self.hidden_size:])**2) + dc_next_b
            dit = dc * cct * it * (1 - it)
            dft = dc * self.cache['backward'][t+1][1] if t < T-1 else dc * 0
            dft = dft * ft * (1 - ft)
            dcct = dc * it * (1 - cct**2)
            
            # Update parameter gradients
            dWf_b += np.dot(dft, concat.T)
            dWi_b += np.dot(dit, concat.T)
            dWc_b += np.dot(dcct, concat.T)
            dWo_b += np.dot(do, concat.T)
             
            dbf_b += dft
            dbi_b += dit
            dbc_b += dcct
            dbo_b += do
            
            # Compute gradients for next step
            dconcat = (np.dot(self.backward_lstm.Wf.T, dft) + 
                      np.dot(self.backward_lstm.Wi.T, dit) + 
                      np.dot(self.backward_lstm.Wc.T, dcct) + 
                      np.dot(self.backward_lstm.Wo.T, do))
            
            dh_next_b = dconcat[:self.hidden_size]
            dc_next_b = dc * ft
        
        # Update parameters
        self.Wy -= learning_rate * dWy
        self.by -= learning_rate * dby
        
        # Update forward LSTM parameters
        self.forward_lstm.Wf -= learning_rate * dWf_f
        self.forward_lstm.Wi -= learning_rate * dWi_f
        self.forward_lstm.Wc -= learning_rate * dWc_f
        self.forward_lstm.Wo -= learning_rate * dWo_f
        self.forward_lstm.bf -= learning_rate * dbf_f
        self.forward_lstm.bi -= learning_rate * dbi_f
        self.forward_lstm.bc -= learning_rate * dbc_f
        self.forward_lstm.bo -= learning_rate * dbo_f
        
        # Update backward LSTM parameters
        self.backward_lstm.Wf -= learning_rate * dWf_b
        self.backward_lstm.Wi -= learning_rate * dWi_b
        self.backward_lstm.Wc -= learning_rate * dWc_b
        self.backward_lstm.Wo -= learning_rate * dWo_b
        self.backward_lstm.bf -= learning_rate * dbf_b
        self.backward_lstm.bi -= learning_rate * dbi_b
        self.backward_lstm.bc -= learning_rate * dbc_b
        self.backward_lstm.bo -= learning_rate * dbo_b
    
    def save_model(self, filename):
        """Save the model to files (both weights and scaler)"""
        folder = os.path.dirname(filename)
        if folder != '':
            os.makedirs(folder, exist_ok=True)
        
        # Save the model weights
        model_data = {
            'input_size': self.input_size,
            'hidden_size': self.hidden_size,
            'forward_lstm': {
                'Wf': self.forward_lstm.Wf,
                'Wi': self.forward_lstm.Wi,
                'Wc': self.forward_lstm.Wc,
                'Wo': self.forward_lstm.Wo,
                'bf': self.forward_lstm.bf,
                'bi': self.forward_lstm.bi,
                'bc': self.forward_lstm.bc,
                'bo': self.forward_lstm.bo,
            },
            'backward_lstm': {
                'Wf': self.backward_lstm.Wf,
                'Wi': self.backward_lstm.Wi,
                'Wc': self.backward_lstm.Wc,
                'Wo': self.backward_lstm.Wo,
                'bf': self.backward_lstm.bf,
                'bi': self.backward_lstm.bi,
                'bc': self.backward_lstm.bc,
                'bo': self.backward_lstm.bo,
            },
            'Wy': self.Wy,
            'by': self.by,
        }
        
        # Save the scaler separately
        scaler_data = {
            'scaler_mean': self.scaler.mean_,
            'scaler_scale': self.scaler.scale_,
            'fitted_scaler': self.fitted_scaler
        }
        
        with open(filename, 'wb') as f:
            pickle.dump({'model': model_data, 'scaler': scaler_data}, f)
    
    @staticmethod
    def load_model(filename):
        """Load a model from file"""
        with open(filename, 'rb') as f:
            data = pickle.load(f)
        
        model_data = data['model']
        scaler_data = data['scaler']
        
        # Recreate the model
        model = BiLSTM(model_data['input_size'], model_data['hidden_size'])
        
        # Restore weights
        model.forward_lstm.Wf = model_data['forward_lstm']['Wf']
        model.forward_lstm.Wi = model_data['forward_lstm']['Wi']
        model.forward_lstm.Wc = model_data['forward_lstm']['Wc']
        model.forward_lstm.Wo = model_data['forward_lstm']['Wo']
        model.forward_lstm.bf = model_data['forward_lstm']['bf']
        model.forward_lstm.bi = model_data['forward_lstm']['bi']
        model.forward_lstm.bc = model_data['forward_lstm']['bc']
        model.forward_lstm.bo = model_data['forward_lstm']['bo']
        
        model.backward_lstm.Wf = model_data['backward_lstm']['Wf']
        model.backward_lstm.Wi = model_data['backward_lstm']['Wi']
        model.backward_lstm.Wc = model_data['backward_lstm']['Wc']
        model.backward_lstm.Wo = model_data['backward_lstm']['Wo']
        model.backward_lstm.bf = model_data['backward_lstm']['bf']
        model.backward_lstm.bi = model_data['backward_lstm']['bi']
        model.backward_lstm.bc = model_data['backward_lstm']['bc']
        model.backward_lstm.bo = model_data['backward_lstm']['bo']
        
        model.Wy = model_data['Wy']
        model.by = model_data['by']
        
        # Restore scaler
        model.scaler.mean_ = scaler_data['scaler_mean']
        model.scaler.scale_ = scaler_data['scaler_scale']
        model.fitted_scaler = scaler_data['fitted_scaler']
        
        return model


def prepare_data(data, window_size=7):
    X, y = [], []
    for i in range(len(data) - window_size):
        X.append(data[i:i+window_size])
        y.append(data[i+window_size])
    return np.array(X), np.array(y)


def train_model(model, X_train, y_train, epochs=100, learning_rate=0.005):
    losses = []
    for epoch in range(epochs):
        epoch_loss = 0
        for i in range(len(X_train)):
            x = X_train[i]
            y_true = y_train[i]
            
            # Forward pass
            y_pred, h_combined = model.forward(x)
            
            # Compute loss (MSE)
            loss = (y_pred - y_true)**2
            epoch_loss += loss[0,0]
            
            # Backward pass
            model.backward(x, y_true, y_pred, h_combined, learning_rate)
        
        epoch_loss /= len(X_train)
        losses.append(epoch_loss)
        if epoch % 10 == 0:
            print(f"Epoch {epoch}, Loss: {epoch_loss:.6f}")
    
    return losses


def predict(model, X_test):
    predictions = []
    for x in X_test:
        y_pred, _ = model.forward(x)
        predictions.append(y_pred[0,0])
    return np.array(predictions)


def evaluate(y_true, y_pred):
    # Calculate RMSE
    rmse = np.sqrt(np.mean((y_pred - y_true)**2))
    
    # MAE
    mae = mean_absolute_error(y_true, y_pred)
    
    # MAPE (avoid division by zero)
    mape = np.mean(np.abs((y_true - y_pred) / np.clip(y_true, 1e-8, None))) * 100
    
    # R² Score
    r2 = r2_score(y_true, y_pred)
    
    return rmse, mae, mape, r2


if __name__ == "__main__":
    # Load CSV and extract BTC_Close_BTC as target
    df = pd.read_csv("merged_lstm_vmd.csv")

    # Ensure the column exists
    if 'BTC_Close_BTC' not in df.columns:
        raise ValueError("Column 'BTC_Close_BTC' not found in the dataset")

    # Drop NA values just in case
    btc_prices = df['BTC_Close_BTC'].dropna().values.reshape(-1, 1)
    
    # Initialize and fit StandardScaler
    model = BiLSTM(input_size=1, hidden_size=64)
    model.scaler.fit(btc_prices)
    model.fitted_scaler = True
    
    # Scale the data
    btc_prices_scaled = model.scaler.transform(btc_prices).flatten()
    
    # Prepare data with sliding window
    window_size = 7  
    X, y = prepare_data(btc_prices_scaled, window_size)
    
    # Split into train and test (80/20)
    split = int(0.8 * len(X))
    X_train, y_train = X[:split], y[:split]
    X_test, y_test = X[split:], y[split:]
    
    # Train the model
    print("Training model...")
    losses = train_model(model, X_train, y_train, epochs=100, learning_rate=0.005)
    
    # Make predictions
    print("Making predictions...")
    predictions_scaled = predict(model, X_test)
    
    # Inverse transform the predictions and test data
    predictions = model.scaler.inverse_transform(predictions_scaled.reshape(-1, 1)).flatten()
    y_test_original = model.scaler.inverse_transform(y_test.reshape(-1, 1)).flatten()
    
    # Evaluate model
    rmse, mae, mape, r2 = evaluate(y_test_original, predictions)
    
    # Print results
    print("\nEvaluation Results:")
    print(f"Test RMSE: {rmse:.6f}")
    print(f"Test MAE: {mae:.6f}")
    print(f"Test MAPE: {mape:.2f}%")
    print(f"Test R² Score: {r2:.6f}")
    
    # Save the trained model
    print("\nSaving model...")
    model.save_model("manual_bilstm_btc_model.h5")
    print("Model saved successfully as 'manual_bilstm_btc_model.h5'")
```

### SVR

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# 1. Load dataset
df = pd.read_csv('df_merged_cleaned.csv', parse_dates=['Date'])
df = df.sort_values('Date').reset_index(drop=True)

# 2. Define features and target
features = [
    'BTC_High_BTC', 'BTC_Low_BTC', 'BTC_Open_BTC','GSPC_High_GSPC', 'GSPC_Close_GSPC', 'GSPC_Open_GSPC',
    'GSPC_Low_GSPC', 'XAU_Low_XAU', 'XAU_Open_XAU', 'XAU_Close_XAU',
    'XAU_High_XAU', 'XAU_Volume_XAU'
]
target = 'BTC_Close_BTC'

X_raw = df[features].values
y_raw = df[target].values
dates = df['Date']

# 3. Split data
split_idx = int(len(df) * 0.8)
X_train_raw, X_test_raw = X_raw[:split_idx], X_raw[split_idx:]
y_train_raw, y_test_raw = y_raw[:split_idx], y_raw[split_idx:]
dates_train, dates_test = dates[:split_idx], dates[split_idx:]

# 4. Standardize features and target
X_scaler = StandardScaler()
y_scaler = StandardScaler()

X_train = X_scaler.fit_transform(X_train_raw)
X_test = X_scaler.transform(X_test_raw)

y_train = y_scaler.fit_transform(y_train_raw.reshape(-1, 1)).flatten()
y_test = y_scaler.transform(y_test_raw.reshape(-1, 1)).flatten()

# 5. SVR implementation
class SVR:
    def __init__(self, C=10, epsilon=0.1, learning_rate=0.0005, epochs=200, batch_size=64, random_state=None):
        self.C = C
        self.epsilon = epsilon
        self.lr = learning_rate
        self.epochs = epochs
        self.batch_size = batch_size
        self.random_state = random_state
        self.weights = None
        self.bias = None
        if self.random_state:
            np.random.seed(self.random_state)

    def _loss(self, y_true, y_pred):
        error = y_true - y_pred
        loss_part = np.maximum(0, np.abs(error) - self.epsilon)
        reg_part = 0.5 * np.dot(self.weights, self.weights)
        return reg_part + self.C * np.sum(loss_part)

    def _gradient(self, X_batch, y_batch):
        n_samples = X_batch.shape[0]
        y_pred = self._predict_batch(X_batch)
        error = y_batch - y_pred
        grad_loss_w = np.zeros_like(self.weights)
        grad_loss_b = 0
        loss_indices = np.abs(error) > self.epsilon
        if np.any(loss_indices):
            sign_error = -np.sign(error[loss_indices])
            grad_loss_w = np.dot(X_batch[loss_indices].T, sign_error)
            grad_loss_b = np.sum(sign_error)
        grad_reg_w = self.weights
        dw = grad_reg_w + self.C * grad_loss_w / n_samples
        db = self.C * grad_loss_b / n_samples
        return dw, db

    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.weights = np.random.randn(n_features) * 0.01
        self.bias = np.random.randn() * 0.01
        for epoch in range(self.epochs):
            indices = np.arange(n_samples)
            np.random.shuffle(indices)
            X_shuffled = X[indices]
            y_shuffled = y[indices]
            for i in range(0, n_samples, self.batch_size):
                X_batch = X_shuffled[i:i+self.batch_size]
                y_batch = y_shuffled[i:i+self.batch_size]
                dw, db = self._gradient(X_batch, y_batch)
                self.weights -= self.lr * dw
                self.bias -= self.lr * db

    def _predict_batch(self, X_batch):
        return np.dot(X_batch, self.weights) + self.bias

    def predict(self, X):
        return np.dot(X, self.weights) + self.bias

# 6. Train the SVR model
svr_model = SVR(C=10, epsilon=0.1, learning_rate=0.0005, epochs=200, batch_size=64, random_state=42)
svr_model.fit(X_train, y_train)

# 7. Predict and inverse transform
y_pred_test_scaled = svr_model.predict(X_test)
y_pred_test = y_scaler.inverse_transform(y_pred_test_scaled.reshape(-1, 1)).flatten()

# 8. Metrics
mae = mean_absolute_error(y_test_raw, y_pred_test)
rmse = np.sqrt(mean_squared_error(y_test_raw, y_pred_test))
r2 = r2_score(y_test_raw, y_pred_test)
mape = np.mean(np.abs((y_test_raw - y_pred_test) / y_test_raw)) * 100

print(f"MAE: {mae:.3f}")
print(f"RMSE: {rmse:.3f}")
print(f"R²: {r2:.3f}")
print(f"MAPE: {mape:.2f}%")

# 9. Results DataFrame
results = pd.DataFrame({
    'Date': dates_test,
    'Actual': y_test_raw,
    'Predicted': y_pred_test
})

# 10. Combine test results with dates
results = pd.DataFrame({
    'Date': dates_test,
    'Actual': y_test_raw,
    'Predicted': y_pred_test
})
print(results.head(10))
```
